Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1.Scatter Plot (scatterplot):
Use: Scatter plots are used to visualize the relationship between two continuous variables. They are often used to identify patterns, correlations, and outliers in the data.

2.Line Plot (lineplot):
Use: Line plots are suitable for visualizing trends and changes in data over a continuous interval, such as time series data.

3.Bar Plot (barplot and countplot):
Use: Bar plots are used to display the distribution of a categorical variable or to compare different categories. countplot is a specific type of bar plot for showing the count of each category.

4.Pair Plot (pairplot):
Use: Pair plots are used to create a grid of scatterplots between all pairs of numerical variables in a dataset. They are useful for identifying relationships and patterns between variables.

5.Histogram (histplot):
Use: Histograms are used to show the distribution of a single variable by dividing it into bins and counting the frequency in each bin. They help visualize data's central tendency and spread.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
```

```python
FMRI = sns.load_dataset("fmri")
```

```python
FMRI
```

```python
sns.lineplot(x = "timepoint" , y = "signal" , data = FMRI ,hue ="event", style = "region" )
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
diamond = sns.load_dataset("diamonds")
```

```python
diamond
```

```python
sns.histplot(x ="price",data= diamond, hue ="cut", element="step" )
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
